```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("shopping-data.csv")
```

```python
df.tail()
```

```python
df.info()
```

```python
df=df.drop_duplicates()
df.isnull().sum()
```

```python
df.dtypes
```

```python
a=["Age","Annual Income (k$)","Spending Score (1-100)"]
```

```python
for i in a:
    print(df[i].describe())
    print(df[i].skew())
    sns.distplot(df[i])
    plt.show()
```

```python
sns.countplot(x="Genre",data=df)
```

```python
data=df.iloc[:,3:5]
data.head()
```

```python
from sklearn.cluster import KMeans
algo=KMeans(n_clusters=2)
algo.fit(data)
```

```python
cen=algo.cluster_centers_
```

```python
algo.labels_
```

```python
algo.inertia_
```

```python
sns.scatterplot(data['Annual Income (k$)'],data['Spending Score (1-100)'],hue=algo.labels_)
sns.scatterplot(cen[:,0],cen[:,1],color='r')
```

```python
dis=[]
k=range(1,15)
for i in k:
    algo=KMeans(n_clusters=i)
    algo.fit(data)
    dis.append(algo.inertia_)
```

```python
dis
```

```python
plt.plot(k,dis)
plt.show()
```

```python
algo1=KMeans(n_clusters=5)
algo1.fit(data)
```
